```python
# %% Importação das Bibliotecas Necessárias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from PyEMD import CEEMDAN
from scipy.signal import hilbert
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Bidirectional, LayerNormalization, MultiHeadAttention, RepeatVector, TimeDistributed, Concatenate, Lambda
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.regularizers import l2

# %% Leitura e Preparação dos Dados
# Substitua "dataset.csv" pelo caminho correto do seu arquivo CSV
data = pd.read_csv("dataset.csv")

# Verificar se a coluna 'id' existe
if 'id' not in data.columns:
    raise ValueError("A coluna 'id' não existe no dataset.")

# Converter 'id' para datetime
data['id'] = pd.to_datetime(data['id'], errors='coerce')
data = data.set_index('id')
data = data.dropna(subset=['ws100'])  # Remover linhas onde 'ws100' é NaN

# Exibir as primeiras linhas para verificação
print("Dados Carregados:")
print(data.head())

# %% Aplicação do CEEMDAN à Série Temporal 'ws100'
ceemdan = CEEMDAN()
ceemdan.trials = 100  # Número de ensaios
ceemdan.noise_width = 0.05  # Intensidade do ruído adicionado

ws100_values = data['ws100'].values
imfs = ceemdan(ws100_values)

# Verificar o número de IMFs gerados
num_imfs = imfs.shape[0]
print(f"Número de IMFs gerados: {num_imfs}")

# Verificar se pelo menos uma IMF foi gerada
if num_imfs < 1:
    raise ValueError("Nenhuma IMF foi gerada pelo CEEMDAN. Verifique os parâmetros ou os dados de entrada.")

# %% Aplicação da Transformada de Hilbert às IMFs
# Calculando a envoltória e a frequência instantânea para cada IMF
# Note que a Transformada de Hilbert requer uma taxa de amostragem
# Defina a taxa de amostragem de acordo com seus dados
# Por exemplo, se os dados são horários, a taxa de amostragem é 1 hora
# Aqui, assumiremos que os dados são horários
sampling_rate = 6  # Ajuste conforme necessário

instantaneous_frequencies = []
instantaneous_amplitudes = []

for i, imf in enumerate(imfs):
    analytic_signal = hilbert(imf)
    amplitude_envelope = np.abs(analytic_signal)
    instantaneous_phase = np.unwrap(np.angle(analytic_signal))
    instantaneous_frequency = np.diff(instantaneous_phase) / (2.0 * np.pi) * sampling_rate
    instantaneous_frequencies.append(instantaneous_frequency)
    instantaneous_amplitudes.append(amplitude_envelope[:-1])  # Ajustar o comprimento

# Converter listas para arrays
instantaneous_frequencies = np.array(instantaneous_frequencies)
instantaneous_amplitudes = np.array(instantaneous_amplitudes)

# Verificar as formas
print(f"Forma das Frequências Instantâneas: {instantaneous_frequencies.shape}")
print(f"Forma das Amplitudes Instantâneas: {instantaneous_amplitudes.shape}")

# %% Preparação das IMFs para o Modelo NAR
# Transpor IMFs para ter cada IMF como uma coluna
imfs_df = pd.DataFrame(imfs.T, index=data.index, columns=[f'IMF_{i+1}' for i in range(num_imfs)])

# Concatenar as IMFs com as outras variáveis, se necessário
# Aqui, consideramos apenas 'ws100' e as IMFs. Você pode adicionar outras variáveis se desejar.
combined_data = pd.concat([data['ws100'], imfs_df], axis=1)

# Verificar as primeiras linhas
print("Dados Combinados com IMFs:")
print(combined_data.head())

# %% Escalonamento dos Dados
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(combined_data)
scaled_data = pd.DataFrame(scaled_data, columns=combined_data.columns, index=combined_data.index)

# Verificar os primeiros valores escalonados
print("Dados Escalonados:")
print(scaled_data.head())

# %% Criação das Sequências para o Modelo NAR
input_steps = 36
output_steps = 6

def create_nar_sequences(data, input_steps, output_steps):
    X, y = [], []
    for i in range(len(data) - input_steps - output_steps + 1):
        X.append(data.iloc[i:(i + input_steps)].values)
        y.append(data.iloc[(i + input_steps):(i + input_steps + output_steps), 0].values)  # 'ws100' está na coluna 0
    return np.array(X), np.array(y)

X, y = create_nar_sequences(scaled_data, input_steps, output_steps)

print(f"Forma de X: {X.shape}")
print(f"Forma de y: {y.shape}")

# %% Divisão em Conjuntos de Treinamento e Teste
split_index = int(X.shape[0] * 0.8)
X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y[:split_index], y[split_index:]

print(f"Forma de X_train: {X_train.shape}")
print(f"Forma de X_test: {X_test.shape}")
print(f"Forma de y_train: {y_train.shape}")
print(f"Forma de y_test: {y_test.shape}")

# %% Definição do Modelo NAR com LSTM e Atenção
def custom_huber_loss(y_true, y_pred):
    delta = 0.2
    error = y_true - y_pred
    abs_error = tf.abs(error)
    quadratic = tf.minimum(abs_error, delta)
    linear = abs_error - quadratic
    loss = tf.reduce_mean(0.5 * tf.square(quadratic) + delta * linear)
    return loss

def build_nar_model(input_steps, num_features, output_steps):
    num_layers = 1
    units = 64
    optimizer = tf.keras.optimizers.Adam(learning_rate=1e-4)
    
    # Encoder
    encoder_inputs = tf.keras.Input(shape=(input_steps, num_features), name='encoder_inputs')
    encoder_outputs = encoder_inputs
    for i in range(num_layers):
        encoder_outputs = Bidirectional(LSTM(
            units=units,
            return_sequences=True,
            kernel_regularizer=l2(1e-5),
            name=f'encoder_lstm_{i}'
        ))(encoder_outputs)
        encoder_outputs = LayerNormalization(name=f'encoder_norm_{i}')(encoder_outputs)
    
    context_vector = encoder_outputs  
    
    decoder_inputs = RepeatVector(output_steps)(encoder_outputs[:, -1, :]) 
    decoder_outputs = decoder_inputs
    for i in range(num_layers):
        decoder_outputs = LSTM(
            units=units * 2,
            return_sequences=True,
            kernel_regularizer=l2(1e-5),
            name=f'decoder_lstm_{i}'
        )(decoder_outputs)
        decoder_outputs = LayerNormalization(name=f'decoder_norm_{i}')(decoder_outputs)
    
    attention_layer = MultiHeadAttention(num_heads=4, key_dim=units, name='multihead_attention')
    attention_output = attention_layer(query=decoder_outputs, value=context_vector, key=context_vector)

    attention_dense = Dense(units, activation='tanh', name='attention_dense')
    attention_output = attention_dense(attention_output)

    decoder_combined_context = Concatenate(axis=-1)([decoder_outputs, attention_output])
    
    outputs = TimeDistributed(Dense(1, activation='linear'), name='output_layer')(decoder_combined_context)
    outputs = Lambda(lambda x: tf.squeeze(x, axis=-1))(outputs)  
    
    model = tf.keras.Model(inputs=encoder_inputs, outputs=outputs, name='Bidirectional_LSTM_Seq2Seq_with_Attention')
    
    model.compile(
        optimizer=optimizer,
        loss=custom_huber_loss,
        metrics=['mae']
    )
    
    return model

# %% Construção do Modelo
num_features = X_train.shape[2]
model = build_nar_model(input_steps, num_features, output_steps)

model.summary()

# %% Definição dos Callbacks
callbacks = [
    EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-6),
    ModelCheckpoint('best_model.h5.keras', monitor='val_loss', save_best_only=True)
]

# %% Treinamento do Modelo
history = model.fit(
    X_train, y_train, 
    epochs=100,  
    batch_size=32, 
    validation_data=(X_test, y_test),
    callbacks=callbacks,
    verbose=1
)

# %% Carregamento do Melhor Modelo
model.load_weights('best_model.h5.keras')

# %% Previsões no Conjunto de Teste
y_pred = model.predict(X_test)

# Verificação das formas
print(f"Forma de y_pred: {y_pred.shape}")

# %% Inversão do Escalonamento para 'ws100'
# Como 'ws100' foi escalado, precisamos reverter o escalonamento para interpretar os resultados
# O 'ws100' está na primeira coluna do scaler
scaler_ws100 = MinMaxScaler()
scaler_ws100.min_ = scaler.min_[0]
scaler_ws100.scale_ = scaler.scale_[0]

# Função para inversão do escalonamento
def inverse_transform_ws100(scaled_values):
    # scaled_values é um array de previsões
    return scaled_values / scaler_ws100.scale_ + scaler_ws100.min_

y_test_inversed = inverse_transform_ws100(y_test)
y_pred_inversed = inverse_transform_ws100(y_pred)

# %% Visualização dos Resultados
output_steps = 6

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('Comparação entre Valores Reais e Previstos para Cada Passo de Previsão', fontsize=16)

for step in range(output_steps):
    row = step // 3
    col = step % 3
    ax = axes[row, col]
    
    real = y_test_inversed[:, step]
    pred = y_pred_inversed[:, step]
    
    ax.plot(real, label='Real', color='blue', alpha=0.7)
    ax.plot(pred, label='Previsto', color='red', alpha=0.7)
    
    ax.set_title(f'Passo de Previsão {step + 1}', fontsize=14)
    ax.set_xlabel('Amostras', fontsize=12)
    ax.set_ylabel('ws100', fontsize=12)
    ax.legend()
    ax.grid(True)
    
plt.tight_layout(rect=[0, 0.03, 1, 0.95])  
plt.show()
```
